```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp
from IPython.display import IFrame

# Funktion fr das Differenzialgleichungssystem
def f(t, z):
    xp = np.array([z[1],-100*(z[1]*z[0]**2-B*z[1]+z[0]-L0)]) #links eigentlich immer z[1], rechts von Komma dann z' Gleichung z[0]=x(t), z[1]=x'(t)
    return xp

#Euler Funktion definieren (immer gleich eigentlich, copy und paste)
def Euler(t0,tn,h,z0):
    n = int((tn-t0)/h) #Die Anzahl der Schritte als Integer
    t = np.linspace(t0,tn,n+1) #Bereich definieren, wo angezeigt werden soll
    z = np.zeros([len(z0),n+1])
    z[:,0] = z0
    for k in range(n):
        z[:,k+1] = z[:,k]+h*f(t[k],z[:,k])
    return t, z

#Parametereingabe
B = 1.0
L0 = 0.5
x_0 = 0.6
xp_0 = 0
t0 = 0.0 #Startwert des Bereichs
tn = 10 #Endwert des Bereichs
h = 0.001 #Schrittweite, hier testen welche am besten Passt
z0 = np.array([x_0,xp_0]) #wichtig, hier Anfangswertprobleme einfgen, sind die Startwerte, diese immer hier eingeben
t, z = Euler(t0,tn,h,z0)
#print('t = ',t)
#print('z = ',z)


# Schaubild-Einstellungen
fig, (ax1) = plt.subplots(1)
ax1.plot(t,z[0,:])
ax1.set(xlabel='$t$', ylabel='$x$')
    
fig, ax = plt.subplots()
ax.plot(z[0,:],z[1,:])
ax.set(xlabel='$x$', ylabel="$\dot{x}$")

plt.show()


```
