```python
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import IFrame

IFrame('Probetestat1_Aufgabe_1.png', width=900, height=720)
```

```python
#-------------Erstellen des Koordinatensystems--------------
    
#Koordinaten Grundgerüst erstellen
x_min, x_max, y_min, y_max = 0, 6.5, -1, 2.5 #erstellen des Bereichs des Koordinatensystems
fig, ax = plt.subplots(figsize=(((x_max-x_min)),(y_max-y_min)))#zuordnen einer Figur und bestimmen wie groß es sein soll
ax.set_aspect('equal') #gleiche Verhälntisse also 1 zu 1, damit Grafik nicht verzerrt wird
ax.axis([x_min,x_max,y_min,y_max]) #anpassen der gleichen Verhältnisse auf gegebenen Bereich

#Haupt-Gitter (Hinterlegen des Gitters)
x_ticks_major = np.arange(x_min, x_max,1) #Variable zur Beschriftung der Skalierung des x-Abschnitts, hier ohne +1, für kleinere Anzeige
y_ticks_major = np.arange(y_min, y_max,1)
ax.set_xticks(x_ticks_major) #dem Achsenabschnitt wird die Skalierung zugeordnet
ax.set_yticks(y_ticks_major)

#Minor-Gitter
x_ticks_minor = np.arange(x_min, x_max, 0.5) #0.5 da minorgrid nicht in 1er Abstand erstellt wird sondern 0.5
y_ticks_minor = np.arange(y_min, y_max, 0.5)
ax.set_xticks(x_ticks_minor,minor=True)#Achsenabschnitt wird die Skalierung zugeordnet
ax.set_yticks(y_ticks_minor,minor=True)

#Gitter anzeigen
ax.grid(which='minor',alpha=0.1)
ax.grid(which='major',alpha=0.5)


#------------Einzeichnen der Punkte und des Graphen------------

#einzeichnen der Punkte als Sterne
points_x = [1,2,3,4,5]
points_y = [0,0,1,0,0]
ax.scatter(points_x, points_y, color='red', marker='*', s=80, label='Einzelne Punkte') #einzeichnen der Punkte, marker='*', s=80
                                                                                       #notwendig für Sterne als Markierung

#----------Berechnen des Polynoms-----------

#Zuweisung der x-Werte für das LGS
x = np.array([1,2,3,4,5])

#Matrix nach Musterschema
A = np.transpose(np.array([x**0,x**1,x**2,x**3,x**4])) #Schreibweise der Matrix
b = np.array([0,0,1,0,0]) #Zuweiseung der y-Werte

#Lösungswerte für x0,x1,x2,x3,x4, löst Matrix auf
a = np.linalg.solve(A,b)

#linspace erzeugt gleichmäßig verteilte Sequenz von Zahlen, zwischen Startwert 0.5 und 5.5 der x-Werte
x = np.linspace(0.5,5.5) #wichtig gibt an wo man den Graph sieht, muss der Bereich sein wo man es sehen soll und kein
                         #markierter Punkt, der für Polynome notwendig ist!

#Alles in LGS einsetzen um Werte für a0, a1, a2, a3, a4 zu erhalten
y = a[0]+a[1]*x+a[2]*x**2+a[3]*x**3+a[4]*x**4

container_x = x
container_y = y

ax.plot(container_x,container_y, '-b')

display(fig)


```
