```python
import numpy as np
import matplotlib.pyplot as plt

x_min, x_max, y_min, y_max = -5, 5, -5, 5
fig, ax = plt.subplots(figsize=(((x_max-x_min)/3),(y_max-y_min)/3))
ax.set_aspect('equal')
ax.axis([x_min, x_max, y_min, y_max])

x_ticks_major= np.arange(x_min, x_max, 1)
y_ticks_major= np.arange(y_min, y_max, 1)
ax.set_xticks(x_ticks_major)
ax.set_yticks(y_ticks_major)

x_ticks_minor= np.arange(x_min, x_max, 0.5)
y_ticks_minor= np.arange(y_min, y_max, 0.5)
ax.set_xticks(x_ticks_minor, minor=True)
ax.set_yticks(y_ticks_minor, minor=True)

ax.grid(which='minor', alpha=0.1)
ax.grid(which='major', alpha=0.5)

'''#Alter falscher Ansatz
t = np.sqrt(zB 3) oder t = np.sqrt(t) #nicht möglich, ansonsten würde es nur einmal verwendet werden
t1 = np.linspace(0, 8*np.pi, 100) <- essenziell wichtig, dieses t zu verwenden! da ja mehrere Werte von t für die Wurzel
genommen werden sollen

x1 = 0+t*np.sin(t1) #falscher Ansatz da t ja eine Wurzel sein soll
y1 = 0+t*np.cos(t1)

ax.plot(x1,y1,'-k')
display(fig)'''

t_values = np.linspace(0, 8 * np.pi, 1000)
x1 = np.sqrt(t_values) * np.sin(t_values)
y1 = np.sqrt(t_values) * np.cos(t_values)
ax.plot(y1, x1, '-k') #tauschen von x1 und y1, sodass Funktion gespiegelt wird 
plt.show()
```
